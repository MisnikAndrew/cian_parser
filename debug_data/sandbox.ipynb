```python
import matplotlib.pyplot as plt
import numpy as np



x_values = [
    55000, 55000, 55000, 55000, 55000, 59000, 60000, 60000, 60000, 60000, 60000, 60000,
    65000, 65000, 65000, 65000, 65000, 65000, 67000, 67900, 68000, 69000, 69000,
    70000, 70000, 70000, 70000, 70000, 75000, 75000, 75000, 75000, 75000, 75000, 77000,
    80000, 80000, 80000, 80000, 80000, 80000, 80000, 80000, 80000, 83300, 85000, 85000,
    85000, 85800, 87000, 90000, 90000, 90000, 90000
]

y_values = [
    645775, 765772, 770615, 776500, 776721, 786768, 690293, 770912, 776330, 779280, 790063, 792086,
    719591, 746702, 771045, 774745, 783167, 802542, 740314, 748655, 757046, 773949, 783178,
    720892, 751539, 766936, 768794, 774489, 722725, 725332, 742783, 756029, 762698, 801642, 745550,
    739271, 743915, 752322, 754641, 757283, 760295, 785042, 788065, 625864, 740210, 777104, 785278,
    749860, 771126, 741979, 757414, 765162, 774648, 784053
]


window_size = 10
moving_avg_y_values = np.convolve(y_values, np.ones(window_size) / window_size, mode='valid')

# For x_values, we use the middle of the window for plotting the moving average
moving_avg_x_values = x_values[(window_size - 1) // 2:-(window_size // 2)]

# Create the plot
plt.figure(figsize=(14, 8))

# Scatter plot
plt.scatter(x_values, y_values, c='blue', marker='o', label='Data')

# Moving average plot
plt.plot(moving_avg_x_values, moving_avg_y_values, c='red', linewidth=2, label=f'{window_size}-Point Moving Average')

# Add labels and title
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot of X vs Y with Moving Average')

# Show grid
plt.grid(True)

# Add legend
plt.legend()

# Display the plot
plt.show()
```
